```python
import os
import json
import requests

API_KEY = os.getenv("OPENROUTER_API_KEY", "api-key")
URL = "https://openrouter.ai/api/v1/chat/completions"
HEADERS = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

messages = [{"role": "system", "content": "Ğ¢Ñ‹ Ğ¿Ğ¾Ğ»ĞµĞ·Ğ½Ñ‹Ğ¹ Ğ°ÑÑĞ¸ÑÑ‚ĞµĞ½Ñ‚."}]
print("Ğ”Ğ¸Ğ°Ğ»Ğ¾Ğ³ Ñ OpenRouter (ÑÑ‚Ñ€Ğ¸Ğ¼Ğ¸Ğ½Ğ³). Ğ’Ğ²ĞµĞ´Ğ¸ 'exit' Ğ´Ğ»Ñ Ğ²Ñ‹Ñ…Ğ¾Ğ´Ğ°.\n")

while True:
    user_text = input("> ")
    if user_text.strip().lower() == "exit":
        break

    messages.append({"role": "user", "content": user_text})

    payload = {
        "model": "deepseek/deepseek-chat-v3.1",
        "messages": messages,
        "stream": True  #
    }

    try:
        resp = requests.post(URL, headers=HEADERS, json=payload, stream=True, timeout=60)
        resp.raise_for_status()

        assistant_chunk = ""  # ÑĞ¾Ğ±ĞµÑ€Ñ‘Ğ¼ Ğ¿Ğ¾Ğ»Ğ½Ñ‹Ğ¹ Ğ¾Ñ‚Ğ²ĞµÑ‚
        print("assistant: ", end="", flush=True)

        for line in resp.iter_lines(decode_unicode=True):
            if line.startswith("data: "):
                data_str = line[6:]
                if data_str == "[DONE]":
                    break
                try:
                    chunk = json.loads(data_str)
                    delta = chunk["choices"][0]["delta"]
                    if "content" in delta:
                        piece = delta["content"]
                        print(piece, end="", flush=True)
                        assistant_chunk += piece
                except Exception:
                    pass  # Ğ¿Ñ€Ğ¾Ğ¿ÑƒÑĞºĞ°ĞµĞ¼ Ğ½Ğµ-JSON Ğ¸Ğ»Ğ¸ Ğ¾ÑˆĞ¸Ğ±ĞºĞ¸

        print("\n")  # Ğ¿ĞµÑ€ĞµĞ²Ğ¾Ğ´ ÑÑ‚Ñ€Ğ¾ĞºĞ¸ Ğ¿Ğ¾ÑĞ»Ğµ Ğ¾ĞºĞ¾Ğ½Ñ‡Ğ°Ğ½Ğ¸Ñ

        # Ğ”Ğ¾Ğ±Ğ°Ğ²Ğ»ÑĞµĞ¼ Ğ¿Ğ¾Ğ»Ğ½Ñ‹Ğ¹ Ğ¾Ñ‚Ğ²ĞµÑ‚ Ğ² ĞºĞ¾Ğ½Ñ‚ĞµĞºÑÑ‚
        messages.append({"role": "assistant", "content": assistant_chunk})

    except Exception as e:
        print("ĞÑˆĞ¸Ğ±ĞºĞ°:", e)
```
